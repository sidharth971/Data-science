```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'22-04-2026\house-prices-advanced-regression-techniques\train.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['Id'], inplace=True)
df.head()
```

```python
missing_values = df.isnull().sum()
missing_values[missing_values>0].sort_values(ascending=False)
```

```python
mean_missing = df.isnull().mean()
mean_missing
```

```python
miss_threshold = 0.65
mean_missing[mean_missing>miss_threshold].index.to_list()
df.drop(columns=mean_missing[mean_missing>miss_threshold].index.to_list(), inplace=True)
df.shape
```

```python
corr_df = df.select_dtypes(include=np.number).corr()['SalePrice'].abs()
df.drop(columns=corr_df[corr_df < 0.1].index, inplace=True)
df
```
