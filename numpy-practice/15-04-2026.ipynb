```python
import time
import numpy as np
```

```python
a = range(100000000)
b = range(100000000)
start = time.perf_counter()
total_val = [x+y for x, y in zip(a, b)]
end = time.perf_counter()
total_p = end - start
print(total_p)
```

```python
a = np.array(a)
b = np.array(b)
start = time.perf_counter()
total_val = a + b
end = time.perf_counter()
total_n = end - start
print(total_n)
print(f"Numpy is faster than python {total_p/total_n:.2f} percentage")
```
